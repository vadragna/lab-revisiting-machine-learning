```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('../learningSet.csv')
```

```python
data.head()
```

```python
print(data.shape)
data.head(10)
```

```python
data.isna().sum()/len(data)
```

```python
drop_list = []
```

```python
categorical = data.select_dtypes(object)
categorical.head()
```

```python
categorical.isna().sum()/len(categorical)
```

```python
categorical['MAILCODE'] = categorical['MAILCODE'].apply(lambda x: x.replace(" ", "A"))
```

```python
categorical = categorical.apply(lambda x: x.replace(" ", np.NaN))
```

```python
categorical.isna().sum()/len(categorical)
```

```python
nulls_percent_categorical = pd.DataFrame(categorical.isna().sum()/len(data)).reset_index()

nulls_percent_categorical.columns = ['column_name', 'nulls_percentage']
```

```python
nulls_percent_categorical.head()
```

```python
columns_above_threshold = nulls_percent_categorical[nulls_percent_categorical['nulls_percentage']>0.5]
columns_above_threshold['column_name']
```

```python
columns_above_threshold.shape
```

```python
drop_columns_list = list(columns_above_threshold['column_name'])
print(drop_columns_list)
```

```python
categorical.drop(drop_columns_list, axis=1, inplace = True)
```

```python
categorical['GENDER'].value_counts
```

```python
categorical['GENDER'] = categorical['GENDER'].fillna('F')
```

```python
categorical['GENDER'] = [x if x in ('M', 'F') else 'other' for x in categorical['GENDER']]
```

```python
categorical['GENDER'].value_counts()
```

```python
df = pd.DataFrame(categorical['STATE'].value_counts()).reset_index()

df.columns = ['state', 'count']
other_states = list(df[df['count']<2500]['state'])
#other_states
def clean_state(x):
    if x in other_states:
        return 'other'
    else:
        return x
    
categorical['STATE'] = list(map(clean_state, categorical['STATE']))
```

```python
categorical['DOMAIN'] = categorical['DOMAIN'].fillna('R2')
```

```python
# Splitting Domain in two parts

categorical['DOMAIN_A'] = list(map(lambda x: x[0], categorical['DOMAIN']))
categorical['DOMAIN_B'] = list(map(lambda x: x[1], categorical['DOMAIN']))
```

```python
categorical = categorical.drop(columns=['DOMAIN'])
```

```python
categorical['MDMAUD_A'].value_counts()
```

```python
drop_list = drop_list + ['MDMAUD_R', 'MDMAUD_F','MDMAUD_A']
```

```python
# Replacing NaNs in cluster with its mode

cluster_mode = categorical['CLUSTER'].mode()[0]

cluster_mode
```

```python
categorical['CLUSTER'] = categorical['CLUSTER'].fillna(cluster_mode) 
```

```python
categorical['HOMEOWNR'].value_counts(dropna=False)
```

```python
categorical['HOMEOWNR'] = categorical['HOMEOWNR'].fillna('U') # assumption: NAN also means 'we don't know'
```

REMOVING COLUMNS WITH SIMILAR INFORMATION

```python
# We will keep the column RFA_2X. We will delete rest of the columns 

for col_name in categorical.columns:
    if "RFA" in col_name:
        drop_list.append(col_name)   
```

```python
drop_list.remove('RFA_2R')
drop_list.remove('RFA_2A')
drop_list
```

```python
# categorical = categorical.drop(columns=drop_list)
# categorical.head()
```

```python
categorical.isna().sum()
```

```python
categorical = categorical.drop(columns=drop_list)
categorical.head()
```

```python
categorical = categorical.drop(columns=['OSOURCE', 'ZIP'])
categorical.head()
```

```python
categorical.isna().sum()
```

```python
# Age flag is not relevant

categorical = categorical.drop(columns=['AGEFLAG'])
```

```python
categorical.isna().sum()
```

```python
categorical['DATASRCE'].value_counts(dropna = False)
```

```python
categorical['DATASRCE'] = categorical['DATASRCE'].fillna('Unknow')
```

```python
categorical.isna().sum()
```

```python
categorical['GEOCODE2'].mode()[0]
```

```python
categorical['GEOCODE2'] = categorical['GEOCODE2'].fillna(categorical['GEOCODE2'].mode()[0])
```

```python
categorical.to_csv('categorical7_02.csv')
```
